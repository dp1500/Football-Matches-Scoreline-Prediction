```python
top_seven = ['Chelsea','Man United','Liverpool','Tottenham','Man City','Arsenal','Leicester']
```

```python
mid_table = ['West Ham','Southampton','Newcastle','Crystal Palace','Burnley','Everton']
```

```python
not_bottom = ['Chelsea','Man United','Liverpool','Tottenham','Man City','Arsenal','Leicester','West Ham','Southampton','Newcastle','Crystal Palace','Burnley','Everton']
```

```python
def home_avg_against_topSeven(team):
    (sum, count) = (0,0)
    for i in range(len(df2)):
        if df2.iloc[i,0] == team:
            if df2.iloc[i,1] in top_seven:
                sum = sum + df2.iloc[i,2]
                count = count + 1
    return (sum/count)

#returns team's(argument) average goals scored against top seven premeir league teams when playing at home
```

```python
home_avg_against_topSeven('Burnley')
```

```python
def away_avg_against_topSeven(team):
    (sum, count) = (0,0)
    for i in range(len(df2)):
        if df2.iloc[i,1] == team:
            if df2.iloc[i,0] in top_seven:
                sum = sum + df2.iloc[i,3]
                count = count + 1
    return (sum/count)
```

```python
away_avg_against_topSeven('Burnley')
```

```python
def home_avg_against_midTable(team):
    (sum, count) = (0,0)
    for i in range(len(df2)):
        if df2.iloc[i,0] == team:
            if df2.iloc[i,1] in mid_table:
                sum = sum + df2.iloc[i,2]
                count = count + 1
    return (sum/count)
```

```python
home_avg_against_midTable('Liverpool')
```

```python
def away_avg_against_midTable(team):
    (sum, count) = (0,0)
    for i in range(len(df2)):
        if df2.iloc[i,1] == team:
            if df2.iloc[i,0] in mid_table:
                sum = sum + df2.iloc[i,3]
                count = count + 1
    return (sum/count)
```

```python
away_avg_against_midTable('Liverpool')
```

```python
def home_avg_against_bottom(team):
    (sum, count) = (0,0)
    for i in range(len(df2)):
        if df2.iloc[i,0] == team:
            if df2.iloc[i,1] not in not_bottom:
                sum = sum + df2.iloc[i,2]
                count = count + 1
    return (sum/count)
    
```

```python
def away_avg_against_bottom(team):
    (sum, count) = (0,0)
    for i in range(len(df2)):
        if df2.iloc[i,1] == team:
            if df2.iloc[i,0] not in not_bottom:
                sum = sum + df2.iloc[i,3]
                count = count + 1
    return (sum/count)
```

```python
away_avg_against_bottom('Liverpool')
```
